```python
import matplotlib.pyplot as plt
import numpy as np

#0부터 10까지 0.1간격으로 숫자생성
x=np.linspace(0,10,100)
y=np.sin(x)

#그래프 그리기
plt.figure(figsize=(8,4))
plt.plot(x,y,label='Sine Ware',color='blue')
plt.title("simple data visualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
